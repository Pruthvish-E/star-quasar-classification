```python
# 1) Data Cleaning
```

```python
# All the NANâ€™s for categorical columns  replaced with its previous row value
```

```python
df['class'].ffill(inplace=True)
```

```python
df['run'].ffill(inplace=True)
```

```python
df['rerun'].ffill(inplace=True)
```

```python
# All the NANâ€™s for numeric columns  replaced with average of the column
```

```python
df['i'].fillna(df['i'].mean(),inplace=True)
```

```python
# Number of NAN values after filling with respective values
```

```python
# 2)  Normalization and Standardization
```

```python
# Why is normalization important?
''' The normal distribution is important because of the Central limit theorem.  In simple terms, if you have many independent variables that may be generated by all kinds of distributions, assuming that nothing too crazy happens, the aggregate of those variables will tend toward a normal distribution.  This universality across different domains makes the normal distribution one of the centerpieces of applied mathematics and statistics.

Another corollary is that the normal distribution makes math easy - things like calculating moments, correlations between variables, and other calculations that are domain specific.  For that reason, even if a distribution isn't actually normal, it is useful to assume that it is normal to get a good, first-order understanding of a set of data.'''
```

```python
# How does it affect dataset?
'''Normalization is performed on data to remove amplitude variation and only focus on the underlying distribution shape'''

''' normalization brings any dataset to a comparable range. It could be to squash down the data to fit between the range of [0,1] or [-1,1] or anything else!'''
```

```python
#  Different graphs used to check whether the data is normal
# Histograms
# boxplots
#normal probability plot or qq plot or quantile-quantile plot
```

```python
# Normalizing all the numeric columns, to make mean 0 and variance 1
```

```python
# Normalizing only numeric features
cols_to_norm = ['ra','dec','u','g','r','i','z','redshift','mjd']
df[cols_to_norm] = df[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```
